# Deep Soft Recurrent Q-Network (DSRQN)

This Q-Learning strategy is a “very effective value-based model-free reinforcement learning algorithm”.  It works by successively improving its evaluations of the action-value function q. Since portfolio optimization is a pseudo-continuous action space, exploration is done “via adding noise to the action itself”.  In this algorithm, we will focus on exponential normalization or softmax while the action value q is fitted. The exponential normalization function applies the standard exponential function to each element in the neural network output and normalizes these values by dividing by the sum of all these exponentials.  This ensures the sum of the components of the output is one, which means that our portfolio weights will not be leveraged.

Some of the Code was inspired by a project done at Columbia University: https://github.com/selimamrouni/Deep-Portfolio-Management-Reinforcement-Learning

As taken from their ReadMe file:

### RL Framework

The portfolio manager agent is set-up in the way:
- The <b>state</b> (at time t) is the input matrix and the previous portfolio weights (at time t-1). 
- The <b>action</b> is the vector of investment weight (at time t).
- The <b>reward function</b> is defined such as it is the agent's return minus a baseline’s return (baseline is an equi weighted agent - invest in all the possible stocks in the same way) minus a term proportional to the maximum of the weight (this term is set-up to make the agent avoids to invest fully in one stock)
